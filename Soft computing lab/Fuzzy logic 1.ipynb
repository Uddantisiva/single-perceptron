```python
import numpy as np
```

```python
# define fuzzy set
fuzzy_set_A = np.array([0.1,0.4,0.7,1.0])
fuzzy_set_B = np.array([0.3,0.6,0.8,0.9])
```

```python
fuzzy_union = np.maximum(fuzzy_set_A,fuzzy_set_B)
```

```python
fuzzy_intersection = np.minimum(fuzzy_set_A,fuzzy_set_B)
```

```python
fuzzy_compliment_A = 1 - fuzzy_set_A
```

```python
fuzzy_compliment_B = 1 - fuzzy_set_B
```

```python
print("fuzzy union is : ",fuzzy_union)
print("Fuzzy intersection  is  : ",fuzzy_intersection)
print("fuzzy compliment of A is : ",fuzzy_compliment_A)
```

# Fuzzy 2

```python
import numpy as np
import skfuzzy as fuzzy 
import matplotlib.pyplot as plt
```

```python
x=np.arange (0,11)
print("crisp set x: " ,x)
```

```python
A = fuzzy.trimf(x,[0,0,5])
B = fuzzy.trimf(x,[5,10,15])
print("fuzzy set: ",A)
print("fuzzy set : ",B)
```

```python
# Not oprator
A_not = 1 - A
print("not_A :  " ,A_not)
# And operator
A_and_B = np.fmin(A,B)
print("intersection : ", A_and_B)
# Or operator
A_or_B = np.fmax(A,B)
print("union : ",A_or_B)
```

```python
# set of A
plt.figure(figsize=(10,8))
plt.subplot (2,2,1)
plt.plot(x,A,label = "set A",color = "red")
plt.title("Fuzzy Set A")
plt.xlabel("x")
plt.ylabel("Membership Degree")
# set of B
plt.figure(figsize=(10,8))
plt.subplot (2,2,2)
plt.plot(x,B,label = "set B",color = "purple")
plt.title("Fuzzy Set B")
plt.xlabel("x")
plt.ylabel("Membership Degree")
# set of not A
plt.figure(figsize=(10,8))
plt.subplot (2,2,3)
plt.plot(x,A_not,label = "Not A",color = "green")
plt.title("Fuzzy Set not A")
plt.xlabel("x")
plt.ylabel("Membership Degree")
# set of A or B
plt.figure(figsize=(10,8))
plt.subplot (2,2,4)
plt.plot(x,A_or_B,label = "set A or B",color = "black")
plt.title("Fuzzy Set A or B")
plt.xlabel("x")
plt.ylabel("Membership Degree")
# set of A and B
plt.figure(figsize=(10,8))
plt.subplot (2,3,5)
plt.plot(x,A_and_B,label = "set A and B",color = "b")
plt.title("Fuzzy Set A and B")
plt.xlabel("x")
plt.ylabel("Membership Degree")
```
