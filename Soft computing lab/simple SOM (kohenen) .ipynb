```python
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt

# Sample data (e.g., 2D points)
data = np.random.rand(100, 3) # 100 data points with 3 features

# Initialize the SOM with grid size 10x10 and input vector size of 3
som = MiniSom(x=10, y=10, input_len=3, sigma=1.0, learning_rate=0.5)

# Randomly initialize the weights of the SOM
som.random_weights_init(data)

# Train the SOM for 100 iterations
som.train_random(data, 100)

# Plotting the results
plt.figure(figsize=(1, 1))
for i, x in enumerate(data):
    w = som.winner(x) # Get the winning neuron
    plt.text(w[0] + 0.5, w[1] + 0.5, str(i), color='red', fontdict={'size': 12})
plt.title('Self-Organizing Map (SOM)')
plt.show()
```
